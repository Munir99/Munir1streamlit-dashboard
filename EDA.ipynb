```python
import plotly.express as px
import random

car_brands = ['Toyota', 'Honda', 'Ford', 'BMW', 'Mercedes', 'Chevrolet', 'Tesla', 'Nissan', 'Hyundai', 'Audi']
fuel_types = ['Gasoline', 'Diesel', 'Electric', 'Hybrid']
transmissions = ['Automatic', 'Manual']

num_samples = 500  # Number of cars

car_data = {
    'Brand': [random.choice(car_brands) for _ in range(num_samples)],
    'Fuel Type': [random.choice(fuel_types) for _ in range(num_samples)],
    'Transmission': [random.choice(transmissions) for _ in range(num_samples)],
    'Price': [random.uniform(5000, 60000) for _ in range(num_samples)],  
    'Mileage': [random.uniform(5000, 250000) for _ in range(num_samples)],  
    'Horsepower': [random.uniform(100, 500) for _ in range(num_samples)],  
    'Year': [random.randint(2000, 2023) for _ in range(num_samples)], 
}

# Histogram: Prices
fig1 = px.histogram(x=car_data['Price'], nbins=50, title="Car Price Distribution", labels={'x': 'Price ($)'})
fig1.show()

# Histogram: Mileage
fig2 = px.histogram(x=car_data['Mileage'], nbins=50, title="Car Mileage Distribution", labels={'x': 'Mileage (miles)'})
fig2.show()

# Scatter Plot: Price vs. Mileage
fig3 = px.scatter(x=car_data['Mileage'], y=car_data['Price'], color=car_data['Brand'],
                  title="Car Price vs. Mileage", labels={'x': 'Mileage (miles)', 'y': 'Price ($)'})
fig3.show()

# Scatter Plot: Price vs. Horsepower
fig4 = px.scatter(x=car_data['Horsepower'], y=car_data['Price'], color=car_data['Fuel Type'],
                  title="Car Price vs. Horsepower", labels={'x': 'Horsepower (HP)', 'y': 'Price ($)'})
fig4.show()

# Scatter Plot: Year vs. Price
fig5 = px.scatter(x=car_data['Year'], y=car_data['Price'], color=car_data['Transmission'],
                  title="Car Price vs. Year", labels={'x': 'Year', 'y': 'Price ($)'})
fig5.show()

```
