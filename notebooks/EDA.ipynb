```python
import plotly.express as px
import random

car_brands = ['Toyota', 'Honda', 'Ford', 'BMW', 'Mercedes', 'Chevrolet', 'Tesla', 'Nissan', 'Hyundai', 'Audi']
fuel_types = ['Gasoline', 'Diesel', 'Electric', 'Hybrid']
transmissions = ['Automatic', 'Manual']

num_samples = 500

car_data = {
    'Brand': [random.choice(car_brands) for _ in range(num_samples)],
    'Fuel Type': [random.choice(fuel_types) for _ in range(num_samples)],
    'Transmission': [random.choice(transmissions) for _ in range(num_samples)],
    'Price': [random.uniform(5000, 60000) for _ in range(num_samples)],
    'Odometer': [random.uniform(5000, 250000) for _ in range(num_samples)],  
    'Horsepower': [random.uniform(100, 500) for _ in range(num_samples)],
    'Year': [random.randint(2000, 2023) for _ in range(num_samples)],
}

# Histograms
px.histogram(x=car_data['Price'], nbins=50, title="Car Price Distribution", labels={'x': 'Price ($)'}).show()
px.histogram(x=car_data['Odometer'], nbins=50, title="Odometer (Mileage) Distribution", labels={'x': 'Odometer (miles)'}).show()
px.histogram(x=car_data['Horsepower'], nbins=50, title="Horsepower Distribution", labels={'x': 'Horsepower (HP)'}).show()
px.histogram(x=car_data['Year'], nbins=50, title="Car Model Year Distribution", labels={'x': 'Year'}).show()

# Scatter Plots
px.scatter(x=car_data['Odometer'], y=car_data['Price'], color=car_data['Brand'], 
           title="Price vs. Odometer", labels={'x': 'Odometer (miles)', 'y': 'Price ($)'}).show()
px.scatter(x=car_data['Horsepower'], y=car_data['Price'], color=car_data['Fuel Type'], 
           title="Price vs. Horsepower", labels={'x': 'Horsepower (HP)', 'y': 'Price ($)'}).show()
px.scatter(x=car_data['Year'], y=car_data['Price'], color=car_data['Transmission'], 
           title="Price vs. Model Year", labels={'x': 'Year', 'y': 'Price ($)'}).show()
px.scatter(x=car_data['Odometer'], y=car_data['Horsepower'], color=car_data['Brand'], 
           title="Odometer vs. Horsepower", labels={'x': 'Odometer (miles)', 'y': 'Horsepower (HP)'}).show()
px.scatter(x=car_data['Year'], y=car_data['Odometer'], color=car_data['Brand'], 
           title="Year vs. Odometer", labels={'x': 'Year', 'y': 'Odometer (miles)'}).show()
```
